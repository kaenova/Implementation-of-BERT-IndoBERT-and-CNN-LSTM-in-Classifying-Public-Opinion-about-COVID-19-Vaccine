```python
import pandas as pd
import numpy as np
import typing
```

```python
# Preprocessing Manual Labeling
df_manual = pd.read_csv("/content/shuffle_v3_Gabungan Voting.csv").rename({"Voting" : "labels"}, axis=1)

# Preprocessing General Training Data
def preprocess_general_training(df_in: pd.DataFrame) -> pd.DataFrame:
  df = df_in.copy()
  df = df[(df["labels"] == 1) | (df["labels"] == -1)]
  df = df.reset_index(drop=True)
  df["labels"] = df["labels"].replace(-1, 0)
  df = df.sample(len(df)).reset_index(drop=True)
  return df
  

df_general_nlu = preprocess_general_training(pd.read_csv("/content/IndoNLU.csv"))
df_general_ridlife = preprocess_general_training(pd.read_csv("/content/Ridlife.csv"))
```

```python
def print_labels(df: pd.DataFrame, name:str = None) -> None:
  labels = {}
  label_list = df["labels"].to_list()
  for i in label_list:
    if i not in labels:
      labels[i] = 0
    labels[i] += 1

  print(f"{name if name is not None else 'Labels'}:",labels)

print_labels(df_manual, "Manual Labeling")
print_labels(df_general_nlu, "IndoNLU")
print_labels(df_general_ridlife, "RiDLife")
```

```python
# Create data Training and Validation
from sklearn.model_selection import train_test_split
validation_size = 0.2

total_negatives = 3412 + 2687
total_positives = 6382 + 2479
print(total_negatives, total_positives)

df_train = pd.concat([df_general_nlu, df_general_ridlife])
df_train_sampled = pd.concat([df_train[df_train["labels"] == 1].sample(total_negatives), df_train[df_train["labels"] == 0]\
                              .sample(total_negatives)]).rename(columns={"clean":"text"}).sample(total_negatives*2)\
                              .reset_index(drop=True)

X_train, X_test, y_train, y_test = train_test_split(df_train_sampled[["raw", "text"]], df_train_sampled[["labels"]], test_size=validation_size)

df_train = pd.concat([X_train.reset_index(drop=True), y_train.reset_index(drop=True)], axis=1)
df_test = pd.concat([X_test.reset_index(drop=True), y_test.reset_index(drop=True)], axis=1)
df_train.to_csv("/content/drive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/4. Train, Test/train/train_4.csv", index=False)
df_test.to_csv("/content/drive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/4. Train, Test/validation/validation_4.csv", index=False)
```

```python
# combine with train data

## Undersampling IndoNLU
sample_target_label = 3412
df_nlu_pos = df_general_nlu[df_general_nlu["labels"] == 1].sample(sample_target_label)
df_nlu_neg = df_general_nlu[df_general_nlu["labels"] == 0].sample(sample_target_label)
# Need to rename for models
df_nlu_train = pd.concat([df_nlu_pos, df_nlu_neg]).sample(sample_target_label * 2).reset_index(drop=True).rename(columns={"clean" : "text"})

## Undersampling RidLife
sample_target_label = 2479
df_ridlife_pos = df_general_ridlife[df_general_ridlife["labels"] == 1].sample(sample_target_label)
df_ridlife_neg = df_general_ridlife[df_general_ridlife["labels"] == 0].sample(sample_target_label)
# Need to rename for models
df_ridlife_train = pd.concat([df_ridlife_pos, df_ridlife_neg]).sample(sample_target_label * 2).reset_index(drop=True).rename(columns={"clean" : "text"})
```

```python
# Concat IndoNLU, RidLife, and Manual Train for Training Data
pd.concat([df_nlu_train, df_ridlife_train, df_manual_train]).to_csv("/content/drive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/4.1 Train Test Baru/50 Manual/train.csv", index=False)
df_manual_test.to_csv("/content/drive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/4.1 Train Test Baru/50 Manual/test.csv", index=False)
```
