```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
path_csv_file = "/content/gdrive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Crawling/bayar.csv"

# Creating list to append tweet data 
tweets_list = []
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('lang:id until:2022-03-25 since:2021-01-13 (covid OR corona OR korona OR covid19 OR covid-19) (vaksinasi OR vaksin OR vaksinisasi) (bayar)').get_items()):
  tweets_list.append([tweet.id, tweet.url, tweet.content])
  # print(tweet.content)
  # callbacks
  if i == 0:
    pd.DataFrame(tweets_list, columns=["TweetID", "TweetURL", "tweet"]).to_csv(path_csv_file, index=False)
    continue
  if i % 500 == 0:
    df = pd.read_csv(path_csv_file, engine='python')
    df_temp = pd.DataFrame(tweets_list, columns=["TweetID", "TweetURL", "tweet"])
    frames = [df, df_temp]
    combined = pd.concat(frames)
    combined.drop_duplicates(inplace=True)
    combined.to_csv(path_csv_file, index=False)
    tweets_list = []
    print(f"Data saved {i}")

df = pd.read_csv(path_csv_file, engine='python')
df_temp = pd.DataFrame(tweets_list, columns=["TweetID", "TweetURL", "tweet"])
frames = [df, df_temp]
combined = pd.concat(frames)
combined.drop_duplicates(inplace=True)
combined.to_csv(path_csv_file, index=False)
tweets_list = []
print(f"Data saved {len(combined)}")
```
