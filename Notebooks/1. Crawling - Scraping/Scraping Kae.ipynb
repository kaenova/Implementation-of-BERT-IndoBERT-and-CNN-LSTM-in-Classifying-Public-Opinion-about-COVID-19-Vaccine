```python
import twint
import nest_asyncio
nest_asyncio.apply()
from datetime import datetime
def scrape_by_city(keywords, folders):
    if folders[len(folders) - 1] != '/':
        folders = folders + '/'
        c = twint.Config()
        final_keywords = keywords + ' lang:id'
        c.Search = final_keywords #search keyword
        c.Since = '2021-01-13'
        c.Until = '2022-03-25'
        c.Store_csv = True
        c.Custom['tweet'] = ['id','user_id', 'tweet']
        c.Output = folders + "{}.csv".format(final_keywords)
        c.Hide_output = False
        twint.run.Search(c)
```

```python
scrape_by_city('(covid OR corona OR korona OR covid19 OR covid-19) (vaksinasi OR vaksin OR vaksinisasi) (efektif)', '//content/gdrive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Crawling')
```
