```python
"""
Script ini digunakan untuk menggabungkan beberapa file csv di dalam suatu folder
dan akan mengeluarkan 1 csv gabungan yang diacak.

@params
input_path: string
            contoh: "/path/menuju/folder/banyak/csv"
csv_output_path: string
                 contoh: "/path/menuju/output.csv"
"""

input_path = "/content/gdrive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/2. Clean/Twitter/v3"
csv_output_path = "/content/gdrive/Shareddrives/Riset Sentimen Vaksin COVID: Yaudahlah/Riset/Data/3. Labeling/shuffle_v3.csv"
```

```python
def concat_and_shuffle(input_dir:str, output_file:str) -> None:
  list_file = os.listdir(input_path)
  all_df = []
  for i in list_file:
    cur_file = input_path+f"/{i}"
    if mimetypes.guess_type(cur_file)[0] != "text/csv":
      continue
    all_df.append(pd.read_csv(cur_file))

  final_df = pd.concat(all_df)
  final_df = final_df.sample(len(final_df))
  print("Total data:", len(final_df))
  print("data has been concatenated and has been shuffled and",
        f"outputted to '{csv_output_path}'")
  final_df.reset_index(drop=True).to_csv(output_file)
  
concat_and_shuffle(input_path, csv_output_path)
```
