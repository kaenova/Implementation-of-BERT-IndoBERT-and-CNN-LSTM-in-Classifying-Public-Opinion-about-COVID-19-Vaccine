```python
# Function for generating word clouds
def generate_wordcloud(data,title):
  wc = WordCloud(width=400, height=330, max_words=100,colormap="Dark2",scale=2).generate(data.to_string())
  plt.figure(figsize=(10,8))
  plt.imshow(wc, interpolation='bilinear')
  plt.title(title)
  plt.axis("off")
  plt.show()
```

```python
for i,df in enumerate(data):
  generate_wordcloud(df['raw'],title[i])
  generate_wordcloud(df['clean'],title[i])
  print('-------------------------------------')
```
