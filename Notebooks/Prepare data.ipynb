```python
# change to -1 and 1 from 0 1
def replace_label(df):
  df = df.copy()
  df["labels"] = df["labels"].replace(0,-1)
  return df
```

```python
#remove label netral
def remove_label_netral(df):
  df = df.copy()
  df = df[(df["labels"]==1) | (df["labels"]==-1)]
  df = df.reset_index(drop=True)
  return df
```

```python
df_ridlife_wn = remove_label_netral(df_ridlife)
df_indonlu_wn = remove_label_netral(df_indonlu)
df_labelgab_wn = replace_label(df_labelgab)
```

```python
df_labelgab_wn.sample(10)
```

```python
def print_labels(df: pd.DataFrame, name:str = None) -> None:
  labels = {}
  label_list = df["labels"].to_list()
  for i in label_list:
    if i not in labels:
      labels[i] = 0
    labels[i] += 1

  print(f"{name if name is not None else 'Labels'}:",labels)

print_labels(df_labelgab_wn, "Manual Labeling")
print_labels(df_indonlu_wn, "IndoNLU")
print_labels(df_ridlife_wn, "RiDLife")
```

# **Undersampling**
[link ref](https://machinelearningmastery.com/undersampling-algorithms-for-imbalanced-classification/)

```python
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split
```

```python
undersample = RandomUnderSampler(sampling_strategy='majority')
X_nlu, y_nlu = undersample.fit_resample(X_nlu.to_numpy().reshape(-1, 1), y_nlu)
```

```python
undersample = RandomUnderSampler(sampling_strategy='majority')
X_ridlife, y_ridlife = undersample.fit_resample(X_ridlife.to_numpy().reshape(-1, 1), y_ridlife)
```

```python
data = {
    'clean':X_nlu[:,0].tolist()+X_ridlife[:,0].tolist(),
    'labels':y_nlu.append(y_ridlife)
}
 
# Create DataFrame
df_under = pd.DataFrame(data)
```

```python
# X_train_uder, X_test_uder, y_train_uder, y_test_uder = train_test_split(df_under['clean'], df_under['labels'], test_size=0.25, random_state=0, stratify=df_under['labels'])
```

```python
undersample = RandomUnderSampler(sampling_strategy='majority')
X, y = undersample.fit_resample(X.to_numpy().reshape(-1, 1), y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0, stratify=y)
```
